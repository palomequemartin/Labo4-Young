```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, fftfreq
import metodos
import plotly.express as px
import plotly.graph_objects as go
from scipy import signal

plt.style.use('./figuras.mplstyle')
```

```python
df = pd.read_csv(f'./dinamico largo.csv')

V = df['Tension fotodiodo [V]'].values
tiempo = df['Tiempo [s]'].values
datos_hasta_t = df['Datos hasta tiempo'].values
duracion = df['Duracion [s]'][0]
fs = int(df['Frecuencia de sampleo [Hz]'][0])


N_puntos = int(duracion*fs)
Deltat = 1/fs
```

```python
fig, ax = plt.subplots()
fin = int(0.5e7)
tf = fin/fs
t = np.linspace(0,tf,len(V[:fin]))
# ax.plot(V)
ax.plot(t,V[:fin]);
ax.set_ylabel('Amplitud [V]')
ax.set_xlabel('Tiempo [s]')
```

```python
fig, ax = plt.subplots()
ax.specgram(V[:fin],Fs = fs,NFFT = int(fs));
ax.set_ylim(0,450)
ax.set_ylabel('Frecuencias [Hz]')
ax.set_xlabel('Tiempo [s]')
```
