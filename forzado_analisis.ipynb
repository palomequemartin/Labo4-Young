```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, fftfreq
import metodos

# %matplotlib ipympl
plt.style.use('./figuras.mplstyle')
```

```python
clase = 3
duracion = 20
fs = 2500
n_medicion = 1


df = pd.read_csv(f'./Mediciones/Clase {clase}/forzado_d{duracion}_fs{fs}({n_medicion}).csv')

V = df['Tension fotodiodo [V]'].values
f_gen = df['Frecuencia generador [Hz]'][0]
duracion = df['Duracion [s]'][0]
fs = df['Frecuencia de sampleo [Hz]'][0]


N_puntos = int(duracion*fs)
Deltat = 1/fs



```

```python
cutoff = 5000
V_trimmed = V[:cutoff]
t = np.linspace(0,int(cutoff/Deltat),len(V_trimmed))
fig, ax = plt.subplots(figsize = (16,4))
ax.set_ylabel('Tensin [V]')
# ax.set_ylim((0, 10))
ax.set_xlabel(f'Tiempo [s]')
ax.set_title(f'Forzado con {f_gen}Hz')
ax.plot(t,V_trimmed, '.-')
plt.savefig(f'Figuras/Clase 3/Forzado con {f_gen}Hz.pdf')
plt.show()
```
