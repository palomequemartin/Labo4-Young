```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, fftfreq


# %matplotlib ipympl
plt.style.use('../figuras.mplstyle')
```

MEDICIÃ“N 0 - 20 SEGUNDOS - 2500 FS

```python
clase = 3
duracion = 20
fs = 2500
n_medicion = 0


df = pd.read_csv(f'../Mediciones/Clase {clase}/forzado_d{duracion}_fs{fs}({n_medicion}).csv')

V = df['Tension fotodiodo [V]'].values
f_gen = df['Frecuencia generador [Hz]'][0]
duracion = df['Duracion [s]'][0]
fs = df['Frecuencia de sampleo [Hz]'][0]


N_puntos = int(duracion*fs)
Deltat = 1/fs
```

```python
fig, ax = plt.subplots(figsize = (16,4))
cutoff = None
if cutoff != None:
    V_trimmed = V[:cutoff]
    t = np.linspace(0,int(cutoff*Deltat),len(V_trimmed))
    ax.plot(t,V_trimmed, '.-')
else:
    t = np.linspace(0,duracion,len(V))
    ax.plot(t,V,'.-')


ax.set_ylabel('TensiÃ³n [V]')
ax.set_xlabel(f'Tiempo [s]')
ax.set_title(f'Forzado con {f_gen} Hz')
        
# figura = plt.gcf()
# plt.show()
```

```python
fig.savefig(f'Clase 3/Forzado con {f_gen} Hz - fs {int(fs)} - mediciÃ³n {n_medicion} - duraciÃ³n {int(duracion)}.pdf')
```

DINAMICO - DURACION 60 - FS 5000

```python
clase = 3
duracion = 60
fs = 5000
n_medicion = 0


df = pd.read_csv(f'../Mediciones/Clase {clase}/dinamico_d{duracion}_fs{fs}({n_medicion}).csv')

V = df['Tension fotodiodo [V]'].values
# f_gen = df['Frecuencia generador [Hz]'][0]
duracion = df['Duracion [s]'][0]
fs = df['Frecuencia de sampleo [Hz]'][0]


N_puntos = int(duracion*fs)
Deltat = 1/fs
```

```python
fig, ax = plt.subplots(figsize = (16,4))
cutoff = 10000
if cutoff != None:
    V_trimmed = V[:cutoff]
    t = np.linspace(0,int(cutoff*Deltat),len(V_trimmed))
    ax.plot(t[7500:8500],V_trimmed[7500:8500], '.-')
else:
    t = np.linspace(0,duracion,len(V))
    ax.plot(t,V,'.-')

fontsize = 18

ax.tick_params(axis='both', which='major', labelsize=13)
ax.tick_params(axis='both', which='minor', labelsize=8)
ax.set_ylabel('TensiÃ³n [V]',fontsize = fontsize)
ax.set_xlabel(f'Tiempo [s]',fontsize = fontsize)
ax.set_title(f'Dinamico 60s 5000fs')
```

```python
fig.savefig(f'Clase 3/Dinamico 3 ciclos.pdf')
```

```python
frecuencias = fftfreq(N_puntos, Deltat)[:N_puntos//2]
amplitudes = np.abs(fft(V)[:N_puntos//2])
fig, ax = plt.subplots()
ax.set_xlabel(r'$f$')
ax.set_ylabel('Frecuencia [Hz]')
ax.set_title(f'FFT')
ax.set_ylabel(r'$|\hat V|$')
ax.set_yscale('log')
ax.plot(frecuencias[100:18000], amplitudes[100:18000])
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Frecuencia [Hz]')

NFFT = 25*1000

ax.specgram(V, NFFT= NFFT, Fs=fs)




# ax.set_xlim((0, 30))
ax.set_ylim((0, 300))
ax.set_title(f'NFFT ={NFFT}')
```

```python
fig.savefig(f'Clase 3/Spectrogram Dinamico - fs {int(fs)} - mediciÃ³n {n_medicion} - duraciÃ³n {int(duracion)} - NFFT = {NFFT}.pdf')
```
